## <b>The Time Independent Schr√∂dinger Equation in 1 Dimension: </b><br>
Removing y, z and t leaves us with the 1 dimensional, time independent Schrodinger equation.<br><br>

#$$i \hbar \frac{\partial}{\partial t}\Psi(\mathbf{r},t) = \hat H \Psi(\mathbf{r},t)$$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
h = 1
m = 1
```

```python
a=1
V=0
```

```python
psi=0
dpsi=1
x=0
dx=a*0.001

xlist=[]
psilist=[]

E=123.12

while x<=a:
  #second derivative of psi
  ddpsi=2*m/h**2*(V-E)*psi

  #first derivative of psi
  dpsi=dpsi+ddpsi*dx

  psi=psi=dpsi*dx
  x=x+dx
  xlist.append(x)
  psilist.append(psi)
  plt.plot(xlist,psilist)
```
